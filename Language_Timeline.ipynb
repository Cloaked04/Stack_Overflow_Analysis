```python
#Lame pre-processing.

js=pd.DataFrame()
java=pd.DataFrame()
csharp=pd.DataFrame()
php=pd.DataFrame()
html=pd.DataFrame()
cpp=pd.DataFrame()
css=pd.DataFrame()
sql=pd.DataFrame()
objc=pd.DataFrame()
c=pd.DataFrame()
ruby=pd.DataFrame()
python=pd.DataFrame()
matlab=pd.DataFrame()
plotList=[js,java,csharp,php,python,html,cpp,css,sql,objc,c,ruby,matlab]
```

```python
#Re-formatting dataframe- converting date-time to acceptable format.

i = 0
for item in dfList:
    item.Frequency = 1
    item['CreationDate'] = pd.to_datetime(item['CreationDate'],format='%Y-%m-%d %H:%M:%S')
    item.set_index('CreationDate',inplace = True)
    # AS is year-start frequency
    
    plotList[i]['Frequency'] = item.Frequency.resample('AS').sum()
    i+=1
```

```python
##All lines below are only to format the dataframe in a clear one-data-per-column. 

Tags=["JS", "JAVA", "C#",'PHP','PYTHON','HTML','C++','CSS','SQL','OBJ-C','C','RUBY','MATLAB']
for item in yearList:
    item['Tags']=Tags
```

```python
#Lets get some colors. These are arbitrary colors I picked up from Category20 palette.

color = pd.DataFrame(['#aec7e8',
 '#ff7f0e',
 '#ffbb78',
 '#2ca02c',
 '#98df8a',
 '#d62728',
 '#ff9896',
 '#1f77b4','#c5b0d5','#e377c2','#9467bd','#7f7f7f','#c7c7c7'],columns=['Color'])

color['Tags']=Tags
color.set_index('Tags',inplace=True)
del color.index.name
```

```python
#Pipeline to make the data Bokeh friendly.

i=0
sources = {}
for df in yearList:
    new_df = pd.concat([df,color], axis=1)
    sources['_' + str(yearNum[i])] = ColumnDataSource(new_df)
    i+=1
```

```python
dictionary_of_sources = dict(zip([x for x in yearNum], ['_%s' % x for x in yearNum]))
dictionary_of_sources
```

```python
js_source_array = str(dictionary_of_sources).replace("'", "")
js_source_array
```

```python
#Initialize the plot.

xdr = Range1d(0, 100)
ydr = Range1d(0,15)
plot = Plot(
    x_range=xdr,
    y_range=ydr,
    plot_width=900,
    plot_height=400,
    outline_line_color=None,
    toolbar_location=None, 
    min_border=20,
)
```

```python
#Let's see what it looks like.
show(plot)
```

```python
#Setting axis characteristics. Take care of intervals in line above last two if you change range.

AXIS_FORMATS = dict(
    minor_tick_in=None,
    minor_tick_out=None,
    major_tick_in=None,
    major_label_text_font_size="10pt",
    major_label_text_font_style="normal",
    axis_label_text_font_size="10pt",

    axis_line_color='#AAAAAA',
    major_tick_line_color='#AAAAAA',
    major_label_text_color='#666666',

    major_tick_line_cap="round",
    axis_line_cap="round",
    axis_line_width=1,
    major_tick_line_width=1,
)

xaxis = LinearAxis(ticker=SingleIntervalTicker(interval=20), axis_label="Number of Queries", **AXIS_FORMATS)
yaxis = LinearAxis(ticker=SingleIntervalTicker(interval=2), axis_label="", **AXIS_FORMATS)   
plot.add_layout(xaxis, 'below')
plot.add_layout(yaxis, 'left')
```

```python
#Let's see what it looks like.

show(plot)
```

```python
#Adding year text. Look out for x and y positions.

text_source = ColumnDataSource({'year': ['%s' % yearNum[0]]})
text = Text(x=30, y=5, text='year', text_font_size='150pt', text_color='#EEEEEE')
plot.add_glyph(text_source, text)
```

```python
#Let's see what it looks like.

show(plot)
```

```python
# Add the circle.
renderer_source = sources['_%s' % yearNum[0]]
circle_glyph = Circle( x='Size',y='y',size='Size',fill_color = 'Color', fill_alpha=0.8, 
    line_color='#7c7e71', line_width=0.5, line_alpha=0.5)

circle_renderer = plot.add_glyph(renderer_source, circle_glyph)
```

```python
# Add the hover (only against the circle and not other plot elements). Added extra tool as x's range changes over years
# the changes are drastic.

tooltips = [('Language',"@index"),('Frequency','@Frequency')]
plot.add_tools(HoverTool(tooltips=tooltips, renderers=[circle_renderer]))
```

```python
show(plot) #Finally something.
```

```python
#Working out the legend. Things to what out for - x,y locations and diff. b/w two tags in legend.

text_x = 80
text_y = 12
for index,row in color.iterrows():
    plot.add_glyph(Text(x=text_x, y=text_y, text=[index], text_font_size='10pt', text_color='#666666'))
    plot.add_glyph(Circle(x=text_x - 3, y=text_y+0.3 , fill_color=row.Color, size=10, line_color=None, fill_alpha=0.8))
    text_y = text_y - 0.8

    
```

```python
show(plot) #Let's see what it looks like.
```

```python
# Add the slider.
code = """
    var year = slider.get('value'),
        sources = %s,
        new_source_data = sources[year].get('data');
    renderer_source.set('data', new_source_data);
    text_source.set('data', {'year': [String(year)]});
""" % js_source_array

callback = CustomJS(args=sources, code=code)
slider = Slider(start=yearNum[0], end=yearNum[-1], value=1, step=1, title="Year", callback=callback)
callback.args["renderer_source"] = renderer_source
callback.args["slider"] = slider
callback.args["text_source"] = text_source
```

```python
show(widgetbox(slider)) #Aao, that widget!
```
